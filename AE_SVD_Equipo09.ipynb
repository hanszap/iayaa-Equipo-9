```python
import numpy as np
import pandas as pd

from sklearn.decomposition import TruncatedSVD
```

#**Ejercicio-1.** 

```python
ratings = pd.read_csv('rating_final.csv')
ratings.head()
```

```python
ratings.shape
```

```python
restaurantes = pd.read_csv('geoplaces2.csv', encoding='latin-1')
restaurantes = restaurantes[['placeID','name']]
restaurantes.head()
```

```python
full_df = pd.merge(ratings, restaurantes, on='placeID')
full_df.head(10)
```

```python
full_df.shape
```

```python
Utility_Matrix = full_df.pivot_table(values='rating', index='userID', columns='name', fill_value=0)
Utility_Matrix.head(10)
```

```python
Utility_Matrix.shape
```

#**Ejercicio-2.**

```python
print('Total de elementos de la matriz de utilidad: %d' % (Utility_Matrix.size))
print('Total de elementos diferentes de cero: %d' % (np.count_nonzero(Utility_Matrix)))
print('Porcentaje de elementos diferentes de cero: %.1f%%' % (100 * np.count_nonzero(Utility_Matrix) / Utility_Matrix.size))
```

#**Ejercicio-3.**

```python
X = Utility_Matrix.T     # Tomamos la traspuesta para que los renglones sean las películas y las columnas los usuarios.
X.shape

num_sv = 128

SVD = TruncatedSVD(n_components=num_sv, random_state=1)

resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape

corrMx = np.corrcoef(resultant_matrix, )
corrMx.shape
```

```python
restaurante = 'Subway'

names = Utility_Matrix.columns
names_list = list(names)
id_megusta = names_list.index(restaurante)

corr_recomienda = corrMx[id_megusta]
print(names)
print(id_megusta)
print(corr_recomienda.shape)
print(corr_recomienda)
```

```python
sorted_index_array = np.argsort(corr_recomienda)
sorted_array = corr_recomienda[sorted_index_array]
rslt = sorted_array[-11 : -1][::-1]
best_rest = sorted_index_array[-11 : -1][::-1]
print("{} mejores calificaciones:".format(10),
      rslt)
print("{} mejores restaurantes:".format(10),
      best_rest)
```

```python
sub_reco_rating = list(names[best_rest])
print(sub_reco_rating)
```

#**Ejercicio-4.**

```python
print('Total de elementos de la matriz de utilidad: %d' % (Utility_Matrix[sub_reco_rating].size))
print('Total de elementos diferentes de cero: %d' % (np.count_nonzero(Utility_Matrix[sub_reco_rating])))
print('Porcentaje de elementos diferentes de cero: %.1f%%' % (100 * np.count_nonzero(Utility_Matrix[sub_reco_rating]) / Utility_Matrix[sub_reco_rating].size))
```
